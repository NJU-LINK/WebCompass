```python

from utils.stat_uitils import load_judge_results, calculate_statistics, print_statistics
```

```python
# Smoke-test session produced by scripts/eval.sh edit
# Layout: results_smoke/<task>/<session>/<page_category>/<id>/judge.json
SESSION_PATH = "./results_smoke/edit/claude-opus-4-6-20260205_text_20260418_164548"
PAGE_CATEGORY = "sp"  # sp | mp
TASK_TYPES = ["edit"]
```

```python
all_stats = {}

for task_type in TASK_TYPES:
    results = load_judge_results(SESSION_PATH, task_type, page_category=PAGE_CATEGORY)

    if not results:
        print(f"No judge results found for {task_type} / {PAGE_CATEGORY}")
        continue

    stats = calculate_statistics(results, task_type)
    all_stats[task_type] = stats

    print_statistics(stats, task_type)


if len(all_stats) > 1:
    print(f"\n{'='*80}")
    print("CROSS-TASK SUMMARY")
    print(f"{'='*80}")
    for task_type, stats in all_stats.items():
        print(f"\n{task_type.upper()}:")
        print(f"  Overall Average: {stats['overall_harmonic_mean']:.4f}")
        print(f"  Overall Worst-of-1 Average: {stats['overall_worst_of_1']:.4f}")
        print(f"  Overall Worst-of-5 Average: {stats['overall_worst_of_5']:.4f}")
        print(f"  Overall Worst Half Tasks Average: {stats['overall_worst_half_tasks']:.4f}")
        print(f"  Total Folders: {stats['total_folders']}")
```
